```python
# We extract all the numeric columns(int ,float) in a dataset
numeric_df = df[numeric_col]
numeric_df
```

<U>Using IQR</U>

```python

Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)

IQR = Q3 - Q1
print(IQR)
```

```python
init = False

for col_name in list(numeric_df.columns):

    low  = Q1[col_name] - 1.5 * IQR[col_name]
    high = Q3[col_name] + 1.5 * IQR[col_name]

    query_string = '{} < @low or {} > @high'.format(col_name, col_name)

    outlier_arr_loop = numeric_df.query(query_string).index
    
    if not init:
        outlier_arr = outlier_arr_loop
        init = True
    
    outlier_arr = outlier_arr.union(outlier_arr_loop)
```

```python
len(outlier_arr)
```

# Check the poverty level of all the members of the house

```python
df['idhogar']
```

```python
len(df['idhogar'].unique())
```

```python
df['Target'].unique()
```

```python
df.groupby(['idhogar','Target']).size().unstack().fillna(0)
```

```python
family_level = df.groupby(['idhogar','Target']).size().unstack().reset_index()
family_level
```

```python
col_names_set = ['HouseholdLevel','Level1','Level2','Level3','Level4']

family_level.columns = col_names_set
family_level
```

```python
family_level.fillna(0,inplace =True)
family_level
```

```python
family_level['total_levels'] = (family_level.Level1 
                               +family_level.Level2 
                               +family_level.Level3  
                               +family_level.Level4)
```

```python
count_of_members_at_diff_level = 0

for i,row in family_level.iterrows():
    max_level = np.max([row.Level1 , row.Level2,row.Level3,row.Level4])
    total_levels = row.total_levels
    
    if total_levels>max_level:
        print("Household Level : {} ; number of members in poverty levels are {} {} {} {}".format(row.HouseholdLevel,
                                                                                                 row.Level1,
                                                                                                 row.Level2,
                                                                                                 row.Level3,
                                                                                                 row.Level4))
        count_of_members_at_diff_level = count_of_members_at_diff_level +1
print('Number of families with family members of different poverty level = {}'.format(count_of_members_at_diff_level))        
```

Thus we observe that 85 families have family members with different poverty level

## Predict the accuracy using random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score,classification_report
```

```python
df.head(10)
```

```python
del_cols = ['Id','idhogar']

df.drop(columns = del_cols ,inplace =True)
```

```python
df.shape
```

```python
obj_columns = df.select_dtypes(['object']).columns
obj_columns
```

```python
df['Target'].dtype
```

```python
X = df.drop(['Target'],axis =1)
y = df.Target
```

```python
X.shape
```

```python
y.shape
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Seems like a good fit

## Check the accuracy using random forest with cross validation

```python
from sklearn.model_selection import KFold,cross_val_score,cross_validate
```

```python
kfold = KFold(n_splits=10 , random_state=12 , shuffle =True)
```

```python
results =  cross_val_score(estimator=rf, 
                              X=X, 
                              y=y, 
                              cv=kfold, 
                              scoring='accuracy')


print('K-Fold accuracy scores : \n', results)
print('Mean score : \n', results.mean())
```

Looks like the model is quite well fitted with accuracy of 90 %
